## Question 1 : 코로나 기간 전후, 서울 시민 경제 활동 분석

#### Q1 - 1 : 코로나 전후 서울 시민 연령대별 지출 금액 비교

- 해당 데이터는 서울 시민의 코로나 전과 후의 연령대별 총 소비 금액을 표기하는 자료입니다.


- 해당 데이터를 통해 2019 ~ 2020년을 비교해보건데, **`경제활동에 참여하고 있는 20대 ~ 70대 중`** 40 ~ 50대를 제외하고는 표면적으로 다른 연령대에서는 지출 금액에 큰 차이가 없는 것으로 보입니다.


- 해당 데이터를 근거로 추측컨데, 코로나 여부와는 별개로 경제활동에는 지속적으로 참여를 하고 있기 때문에 소비하는 패턴은 크게 바뀌지 않은 것으로 보이며, 해당 분석 이후에 40 ~ 50대에서 코로나 전후로 어떤 소비가 줄었는지 확인할 필요가 있습니다.

```python
# 코로나 전 소비 금액
before_covid_cost = shinhan_bank[(shinhan_bank["기준년월"]>=201903)
                                &(shinhan_bank["기준년월"]<202003)].groupby("나이")["총소비금액"].sum()

# 코로나 후 소비 금액
after_covid_cost = shinhan_bank[(shinhan_bank["기준년월"]>201909)&
                               (shinhan_bank["기준년월"]<202103)].groupby("나이")["총소비금액"].sum()
```

```python
plt.plot(before_covid_cost)
plt.plot(after_covid_cost)
plt.legend(['Before Covid', 'After Covid'])
plt.xlabel("서울 시민 연령대")
plt.ylabel("1년 총 소비 금액")
plt.title("연령에 따른 서울 시민 1년 총 소비 금액")
plt.show()
```

#### Q1 - 2 : 직장인 & 자영업 여부에 따른 수입 증감 비교

- 그러나 각 연령대별 소비 금액을 비교할 때와는 반대로 업종에 따라 종사하고 있는 일로 분류 하게 되었을 때, 즉 `수입의 출처로 분류`하게 되었을 때 코로나가 사람들에게 미치는 영향을 보다 명확하게 확인할 수 있습니다.


- 아래 데이터는 먼저 코로나 전후, 서울시에 거주하고 있는 자영업자의 매출 변화를 분석한 그래프입니다.

```python
# 코로나 전후 자영업자 매출 비교
before_covid_sales = shinhan_bank[(shinhan_bank["기준년월"]>=201903)
                                &(shinhan_bank["기준년월"]<202003)&
                                  (shinhan_bank["직장인여부"]==0)].groupby("나이")["가맹점매출입금"].sum()
after_covid_sales = shinhan_bank[(shinhan_bank["기준년월"]>201909)
                                &(shinhan_bank["기준년월"]<202103)&
                                  (shinhan_bank["직장인여부"]==0)].groupby("나이")["가맹점매출입금"].sum()
```

아래 데이터를 통해 볼 수 있듯, **자영업자의 경우, 코로나 전 대비 코로나 이후 매출 차가 심각하게 벌어지고 있으며, 특히 40 ~ 60대에서 큰 차이**를 보이고 있습니다.

해당 기간 뉴스를 확인해 보면, 코로나로 인해 사람들이 서울로 출퇴근하는 횟수가 줄고, 재택 근무를 하게 되는 상황과 위생적인 안전 등을 고려해 자영업자들의 업장 방문횟수가 줄게 되면서 수입에 직격탄을 맞게 되는 것을 알 수 있습니다.

추가적으로 신한은행 데이터를 통해 급여 / 매출 차이를 확인했을 때, 자영업자와 직장인의 수입 차이가 크게 차이납니다.

아마도 이는 자영업자의 경우, 결제 대금을 이체 받는 통장이 신한은행만 있는 경우가 아닌 경우가 더 많고, 직장인의 경우 급여 계좌 형태로 이체 시, 급여를 투명하게 추적할 수 있기 때문에 차이가 발생하는 것으로 보입니다.

결과적으로 자영업자는

- `2019 / 2020 매출 비교` : `약 14억 / 약 7억`

직장인은

- `2019 / 2020 급여 비교` : `약 450억 / 약 545억`

으로 확인이 가능합니다.

그러나 데이터 표본 추출로 인한 절대적인 금액 차이는 둘째치더라도 두 종사 계층 사이의 K-양극화 현상이 발생하고 있는 것을 확인할 수 있습니다.

```python
# 자영업자 전체 vs 직장인 임금 격차 비교

before_covid_sales_total = shinhan_bank[(shinhan_bank["기준년월"]>=201903)
                                &(shinhan_bank["기준년월"]<202003)&
                                  (shinhan_bank["직장인여부"]==0)]["가맹점매출입금"].sum()
after_covid_sales_total = shinhan_bank[(shinhan_bank["기준년월"]>201909)
                                &(shinhan_bank["기준년월"]<202103)&
                                  (shinhan_bank["직장인여부"]==0)]["가맹점매출입금"].sum()

before_covid_income_total = shinhan_bank[(shinhan_bank["기준년월"]>=201903)
                                &(shinhan_bank["기준년월"]<202003)&
                                  (shinhan_bank["직장인여부"]==1)]["급여입금"].sum()
after_covid_income_total = shinhan_bank[(shinhan_bank["기준년월"]>201909)
                                &(shinhan_bank["기준년월"]<202103)&
                                  (shinhan_bank["직장인여부"]==1)]["급여입금"].sum()

# 전체 매출 확인 값
# before_covid_sales_total, after_covid_sales_total, before_covid_income_total, after_covid_income_total


# 소득 격차 시각화
sales = [before_covid_sales_total,before_covid_income_total]
salary = [after_covid_sales_total,after_covid_income_total]

df_bar = pd.DataFrame({"자영업자":sales, "직장인":salary}, index=["자영업자", "직장인"])

bar_chart = df_bar.plot.bar(color=["r","c"], title="코로나 전후 노동자 수입 비교")
bar_chart.set_xlabel("노동자 분류")
bar_chart.set_ylabel("수입")
```
