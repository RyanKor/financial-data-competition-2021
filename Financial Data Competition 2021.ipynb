### 6. 분석 배경

- 제태크의 한 분야인 주식 거래에 관련한 데이터는 사회 초년생인 우리에게 삶에 있어 피와 살이 되는 좋은 정보가 될 것이고

- 코로나 발생을 기점으로 20 ~ 30대의 주식 거래 참여 인원의 증가하면서 기존 주식 거래하는 인원들과 다른 거래 패턴을 가질 것이라 가설을 세움

- 무엇보다 미디어에서 자주 언급되는 빚투 (빚내어 투자) & 영끌 (영혼까지 끌어모아 부동산 투자) 인원이 늘어났다는 것을 자주 볼 수 있는데 신한은행에서 제공하는 서울시 시민 데이터 중 금융 & 부동산 담보 대출 금액과 한국투자증권 동학개미 운동을 엮어 이러한 대출을 통한 투자 비율을 분석할 수 있다는 것을 데이터셋을 통해 확인하게 되었다.(비교 분석하여 대출을 통한 투자 비율을 유추하고자 한다.)

- 최종적으론 가공한 데이터를 토대로 사용자들이 구매하는 주식의 테마 & 카테고리 패턴을 분석할 수 있을 것으로 보이며 19년 대비 20년의 투자 패턴을 미디어 보도 자료와 비교 & 대조하면서 경기 변동에 따른 한국인들의 투자 패턴을 예측할 것으로 보인다.


### 7. 분석 기법
